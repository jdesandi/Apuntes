$$\begin{eqnarray}
\dot{x}&=&y\\
\dot{y}&=&-\frac{k}{m}x-\frac{B}{m}y
\end{eqnarray}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
```

```python
k=3.0
m=1.0
B=0.5

def armonico (variables, t):
    x,y=variables
    return [y,-k*x/m-B/m*y]
inicial=[0.8, 0]
tiempo=np.arange(0,20,.01)
resultado=odeint(armonico,inicial, tiempo)
xx, yy=resultado.T
plt.plot(tiempo, xx, c='r',label='Posicion')
plt.plot(tiempo, yy, c='k',label='Velocidad')
plt.legend(loc='best',fontsize=14)
plt.xlabel('tiempo',fontsize=14)
plt.show()
```

```python
from ipywidgets import *
```

```python
def amortiguado (t=0):
    fig=plt.figure()
    ax=fig.add_subplot(1,1,1)
    ax.plot(tiempo,xx,'k-')
    ax.plot(tiempo[t],xx[t],'ro')
    ax.text(4,0.6,'tiempo=%s'%tiempo[t])
    ax.text(4,0.5,'posicion=%s'%xx[t])
    fig.canvas.draw()
t_f=len(tiempo)
interact_manual(amortiguado,t=(0,t_f,1));
```

```python
r=1
def poblacion (x,t):
    return r*x*(1-x)
```

```python
x0=0.05
tt=np.linspace(0,10)
xx=odeint(poblacion,x0,tt)
plt.plot(tt,xx)
plt.xlabel('$t$',fontsize=10)
plt.ylabel('$x$',fontsize=10)
plt.show()
```

```python
for r in np.arange(-1,1,0.1):
    plt.plot(tt,poblacion(x0,tt))
    plt.xlabel('$t$',fontsize=10)
    plt.ylabel('$x$',fontsize=10)
plt.show()
```
